```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cmocean.cm as cmo
```

```python
ds = xr.open_dataset('/Users/joegradone/Desktop/RG_ArgoClim_33pfit_2019_annual.nc',decode_times=False)
ds
```

```python
plt.figure(figsize=(15,10))
ds.ARGO_TEMPERATURE_MEAN[15,:,:].plot(vmin=15,vmax=25)
plt.xlim(260,380)
plt.ylim(-35,35)
```

```python
plt.figure(figsize=(15,10))
ds.ARGO_SALINITY_MEAN[15,:,:].plot(vmin=33,vmax=37.5)
plt.xlim(260,380)
plt.ylim(-35,35)
```

```python
#ds_clouds = ds.where((ds.levels > bot_mod) & (ds.levels < top_mod))

ds_slice = ds.sel(LONGITUDE=296.25)
ds_slice
```

```python
plt.figure(figsize=(15,10))
plt.contourf(ds_slice.LATITUDE,ds_slice.PRESSURE,ds_slice.ARGO_TEMPERATURE_MEAN,levels=np.arange(4,30,1),cmap=cmo.thermal)
plt.xlim(12,20)
plt.ylim(1000,0)
plt.ylabel('Depth [m]',fontsize=18)
plt.xlabel('Latitude',fontsize=18)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=18)
cbar.set_label(label='Temperature',size=18)
plt.gca().tick_params(labelsize=18)
plt.title('ARGO Climatology 2004-2018 along 66°W',fontsize=18)
```

```python
plt.figure(figsize=(15,10))
plt.contourf(ds_slice.LATITUDE,ds_slice.PRESSURE,ds_slice.ARGO_SALINITY_MEAN,levels=np.arange(34.5,37.5,0.1),cmap=cmo.haline)
plt.xlim(12,20)
plt.ylim(1000,0)
plt.ylabel('Depth [m]',fontsize=18)
plt.xlabel('Latitude',fontsize=18)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=18)
cbar.set_label(label='Salinity',size=18)
plt.gca().tick_params(labelsize=18)
plt.title('ARGO Climatology 2004-2018 along 66°W',fontsize=18)
```
