# Plot up some maps of where glider data is

```python
## Imports
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import glob
import pickle
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.feature as cpf
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import seawater as sw
import cmocean.cm as cmo
from scipy import integrate
from scipy.interpolate import griddata
import sys
import gsw
from erddapy import ERDDAP
pd.options.mode.chained_assignment = None  # default='warn'




def get_erddap_dataset(ds_id, server, variables=None, constraints=None, filetype=None):
    """
    Returns a netcdf dataset for a specified dataset ID (or dataframe if dataset cannot be converted to xarray)
    :param ds_id: dataset ID e.g. ng314-20200806T2040
    :param variables: optional list of variables
    :param constraints: optional list of constraints
    :param filetype: optional filetype to return, 'nc' (default) or 'dataframe'
    :return: netcdf dataset
    """
    variables = variables or None
    constraints = constraints or None
    filetype = filetype or 'nc'
    #ioos_url = 'https://data.ioos.us/gliders/erddap'


    e = ERDDAP(server,
               protocol='tabledap',
               response='nc')
    e.dataset_id = ds_id
    if constraints:
        e.constraints = constraints
    if variables:
        e.variables = variables
    if filetype == 'nc':
        try:
            ds = e.to_xarray()
            ds = ds.sortby(ds.time)
        except OSError:
            print('No dataset available for specified constraints: {}'.format(ds_id))
            ds = []
        except TypeError:
            print('Cannot convert to xarray, providing dataframe: {}'.format(ds_id))
            ds = e.to_pandas().dropna()
    elif filetype == 'dataframe':
        #ds = e.to_pandas().dropna()
        ds = e.to_pandas().dropna(how='all')
    else:
        print('Unrecognized filetype: {}. Needs to  be "nc" or "dataframe"'.format(filetype))

    return ds
```

## Make a map

```python
fig, ax = plt.subplots(figsize=(15, 20), subplot_kw=dict(projection=ccrs.Mercator()))

## Map extent
lon1 = -83
lon2 = -59
lat1 = 9
lat2 = 26

ax.set_extent([lon1, lon2, lat1, lat2])

ax.add_feature(cfeature.LAND, edgecolor='black',facecolor='tan')
ax.add_feature(cfeature.COASTLINE, edgecolor='black')

glb = ax.gridlines(draw_labels=True, alpha=0.5, linewidth=.5 )
glb.xlabels_top = glb.ylabels_right = False
glb.xlabel_style = {'size': 22}
glb.ylabel_style = {'size': 22}

ax.plot(tot_transect_df.longitude.values, tot_transect_df.latitude.values,'.',markersize=5,transform=ccrs.PlateCarree(),label='AOML',color='#33BBEE')
ax.plot(ru36_gdf.longitude.values, ru36_gdf.latitude.values,'.',markersize=5,transform=ccrs.PlateCarree(),label='RU36 Anegada',color='#004488')
ax.plot(ru29_gdf.longitude.values, ru29_gdf.latitude.values,'.',markersize=5,transform=ccrs.PlateCarree(),label='RU29 Windwards',color='#CC3311')

leg = ax.legend(loc='lower left', frameon= False,prop={'size': 22, 'weight':'bold'},markerscale=5,handletextpad=-0.25, labelspacing=0.05,borderaxespad=0.5)

plt.savefig('/Users/joegradone/Desktop/All_Carib_Gliders_Map.png', bbox_inches='tight',dpi=300,facecolor='white')
```

## Temperature, Salinity, and Oxygen Plots from North and South of Leewards and Windwards

```python
## Depth grid
depth_grid = np.arange(0,1001,2)

## Preallocate dataset for interpolation
temp = np.empty((len(depth_grid),4))
temp[:] = np.nan
interp_df_aoml = pd.DataFrame(temp,columns=['depth', 'temperature', 'salinity','oxygen'])
interp_df_aoml.depth = depth_grid

for x in np.arange(0,len(depth_grid)-1):
    depth_ind = np.where((tot_transect_df.ctd_pressure >= depth_grid[x]) & (tot_transect_df.ctd_pressure <= depth_grid[x+1]))[0]
    
    interp_df_aoml.temperature.loc[x] = np.nanmedian(tot_transect_df.temperature.values[depth_ind])
    interp_df_aoml.salinity.loc[x] = np.nanmedian(tot_transect_df.salinity.values[depth_ind])
    interp_df_aoml.oxygen.loc[x] = np.nanmedian(tot_transect_df.aanderaa4831_dissolved_oxygen_shifted.values[depth_ind])
```
